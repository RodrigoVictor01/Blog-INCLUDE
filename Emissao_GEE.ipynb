## EMISSÃO DOS GASES DO EFEITO ESTUFA (GEE) NO BRASIL E SUAS ORIGENS.


##Introdução

A emissão dos Gases do Efeito Estufa (GEE) é um dos fatores que mais contribuem para o aquecimento global. A concentração desses gases na atmosfera impede que o calor seja irradiado, aquecendo ainda mais a superfície terrestre, e, consequentemente, causando alterações climáticas em nosso planeta.

Sendo assim, esse estudo tem como objetivo principal avaliar o volume total de emissão de três GEE no Brasil entre o período de 2011 a 2021. Os gases a serem avaliados de forma geral são o CO2 (dióxido de carbono), CH4 (metano) e CO (monóxido de carbono), com foco no Gás do Efeito Estufa que foi mais emitido entre esse intervalo de anos: o CO2, pois o dióxido de carbono é um gás extremamente relevante na contribuição do efeito estufa. Além disso, queremos também avaliar as fontes de emissão desse gás, com a finalidade de compreender algo que vem afetando diretamente o clima e o ambiente em nosso planeta e que muitas vezes não paramos para pensar as possíveis e drásticas consequências que a não mitigação desse problema pode gerar em um futuro próximo.

Por tratar-se de uma enorme escala de volume, por padrão, iremos adotar a unidade de medida de volume de um gás como tonelada (t), a fim de mensurar os dados que serão levantados.

No gráfico acima, conseguimos nitidamente observar que o principal responsável pela emissão de CO2 na atmosfera são as **MUDANÇAS DE USO DA TERRA E FLORESTAS**, que referem-se principalmente às queimadas e ao desmatamento para, geralmente, dar lugar à pecuária e/ou à agropecuária. No processo das queimadas, a eliminação do material orgânico por fogo gera a emissão de gases poluentes para a atmosfera, sendo o CO2 o principal gás a ser liberado por esses processos antrópicos.

O desmatamento de florestas tambem se enquadra nesta categoria e contrubui para um volume maior de emissão de CO2, pois quando há esse processo, ocorre a liberação de grandes quantidades de carbono na forma de CO2 para atmosfera.

Perceba que ao analisar a categoria citada acima, a origem da liberação dos GEE, neste caso, não são a agropecuária e/ou a pecuária. Os processos mais contribuintes para a emissão de CO2 são o desmatamento e queimadas de florestas. Entretando, a agropecuária e pecuária também tem sua parcela de responsabilidade que contribui com a emissão desses gases, como podemos ver também no gráfico acima. Mas o importante é compreender a origem de maior emissão de CO2, que são as **MUDANÇAS DE USO DA TERRA E FLORESTAS**, e não suas finalidades, que uma delas seria dar lugar à terra desmatada para o agronegócio.

## Emissão de CO2 por resíduos entre 2011 e 2021.

````python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np


anos = list(map(str,range(2011, 2022)))
qtdAnos = len(anos)



# Importando o DataFrame

emissaoCO2 = pd.read_csv('/content/drive/MyDrive/PROJETO DADOS PYTHON/EMISSAO-CO2.csv')

# Criando array a partir do DataFrame na linha de index 0

volumeCO2Residuos = np.array(emissaoCO2.iloc[0])

# Removendo valores indesejáveis
volumeCO2Residuos = np.delete(volumeCO2Residuos, 0)
volumeCO2Residuos = np.delete(volumeCO2Residuos, np.size(volumeCO2Residuos)-1)


# Construção do gráfico

fig, ax = plt.subplots()
fig.set_size_inches(20,5)
ax.set_title('Volume de CO2 emitido por resíduos a cada ano entre 2011 e 2021.')
ax.set_xlabel('PERÍODO')
ax.set_ylabel('VOLUME (t)')
ax.bar(anos, volumeCO2Residuos, color = 'brown')



# Função que adiciona o valor do volume dentro de cada barra do gráfico

def addlabels(x,y):
    for i in range(len(x)):
        plt.text(i,y[i]//2,y[i], ha = 'center', color = '#FFF', weight='bold')

addlabels(anos, volumeCO2Residuos)


plt.show()


# Calculando o volume médio da categoria 'RESIDUOS' a partir do index da linha 0
total = emissaoCO2.iloc[0]
media = (total['TOTAL CATEGORIA'])/qtdAnos

# Desvio padrão populacional
stdDeviation = np.std(volumeCO2Residuos)

# Section de informações complementares

maiorEmissao = max(volumeCO2Residuos)
menorEmissao = min(volumeCO2Residuos)


print(f'O maior volume de emissão de CO2 por resíduos foi em {anos[4]}, atingindo um volume de {maiorEmissao} (t)')
print(f'O menor volume de emissão de CO2 por resíduos foi em {anos[7]}, atingindo um volume de {menorEmissao} (t)')
print(f'O volume médio de CO2 emitido por resíduos de {anos[0]} a {anos[10]} é de aproximadamente {media:.1f} (t)')
print(f'Desvio padrão populacional: {stdDeviation:.4f}')
print(f'Relato da medida: ({media:.1f} ± {stdDeviation:.4f}) (t)')

````

O gráfico acima mostra o volume de emissão de CO2 no período de 10 anos de forma individual. No caso deste gráfico, a análise se baseia na emissão de CO2 por resíduos, que comparada a outras categorias, sua participação na emissão de CO2, em escala, é quase que irrisória.

Mas o que são resíduos? Os principais resíduos responsáveis pela emissão do CO2 são: **Resíduos sólidos urbanos, resíduos orgânicos, combustíveis fósseis, resíduos de biomassa e resíduos industriais.**

Os resíduos sólidos urbanos são o lixo produzido nas cidades, composto por materiais orgânicos e inorgânicos.

Resíduos orgânicos se enquadra em restos de alimentos, esterco animal e outros materiais que liberam o dióxido de carbono durante sua decomposição.

Combustíveis fósseis são resíduos que liberam CO2 a partir da queima de combustíveis fósseis, como gás natural, carvão e petróleo.

Os resíduos de biomassa geram a decomposição de resíduos agrícolas, resíduos de processamento de alimentos e também restos de madeira.

Portanto, para cada resíduo, é necessário que haja o descarte destes. Por exemplo, para que haja o descarte de resíduos sólidos urbanos, materiais orgânicos e inorgânicos são transferidos para aterros sanitários e, quando depositados sem tratamento, passa por um processo de decomposição, liberando CO2. Dessa forma, a maior parte da liberação de CO2 por resíduos ocorre pela decomposição deles mesmos.

O aumento no desmatamento se atribui a uma gama de fatores e contextos. Uma das razões para esse aumento é a expansão agrícola, pois a demanda crescente por terras para a agricultura tem levado à derrubada de florestas, comprometendo os biomas brasileiros e contribuindo com a emissão do CO2 e, consequentemente, aumentando o aquecimento global.

O estudo analisa as emissões de CO2 no Brasil, com foco nas mudanças de uso da terra e florestas, setor energético e processos industriais. Destaca-se o desmatamento e as queimadas como principais causas de emissões de CO2, especialmente na região amazônica com a finalidade de suprir o mercado agropecuário.
